### PRIMERA PRUEBA

X e y. division en Train-test

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Target'])
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=33)
```

Comprobamos shapes

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

Escalado (a train para no contaminar los datos de test)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
X_scaled.shape
```

APLICACION DE PCA PARA REDUCIR FEATURES

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=37)
X_pca = pca.fit_transform(X_scaled)
```

Veamos cómo de buenos predictores son las nuevas dimensiones

```python
pca.explained_variance_ratio_.cumsum()
```

```python
pd.Series(pca.explained_variance_ratio_.cumsum(), index=np.arange(1,38))
```

```python

plt.plot(pd.Series(pca.explained_variance_ratio_.cumsum(), index=np.arange(1,38)))
plt.xlabel('number of PCs')
plt.ylabel("cumulative explained ratio")
```

vamos a quedarnos con 21 PCs

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = Pipeline(steps=[
                    ('scaler', StandardScaler()),
                    ('pca', PCA(n_components=21)),
                    ('classifier', RandomForestClassifier(max_depth=5))
])
pipe
```

```python
# predicciones
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print('Accuracy', accuracy_score(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
pipe_gs = Pipeline(steps=[
                    ('scaler', StandardScaler()),
                    ('pca', PCA()),
                    ('classifier', RandomForestClassifier())
])
pipe_gs
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
        'pca__n_components': np.arange(1,10),
        'classifier__max_depth': np.arange(1,6),
        'classifier__min_samples_leaf': [2,4,6]
}

gs = GridSearchCV(pipe_gs, params, cv=10, scoring='accuracy')
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
pipe_final = Pipeline(steps=[
                    ('scaler', StandardScaler()),
                    ('pca', PCA(n_components=21)),
                    ('classifier', RandomForestClassifier(max_depth=5, min_samples_leaf=6))
])
pipe_final
```

```python
pipe_final.fit(X_train,y_train)
```

```python
y_pred_best = pipe_final.predict(X_test)
```

```python
print('Accuracy', accuracy_score(y_test, y_pred_best))
```
